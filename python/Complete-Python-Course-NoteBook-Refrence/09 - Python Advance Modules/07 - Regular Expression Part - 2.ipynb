# Quantifiers

```python
phone = re.search(r'\d{3}-\d{3}-\d{4}', text)
```

```python
phone
```

```python
phone_pattern = re.compile(r'(\d{3})-(\d{3})-(\d{4})')
```

```python
phone_pattern
```

```python
number = re.search(phone_pattern, text)
```

```python
number.group()
```

```python
number.group(3)
```
